```python
from ekphrasis.classes.preprocessor import TextPreProcessor
from ekphrasis.classes.tokenizer import SocialTokenizer
from ekphrasis.dicts.emoticons import emoticons

text_processor = TextPreProcessor(
    # istilah-istilah yang akan dinormalisasi
    normalize=['email', 'percent', 'money', 'phone', 'user',
        'time', 'date', 'number'],
    annotate={"hashtag"},
    fix_html=True,
    
    # korpus yang digunakan untuk statistik kata yang akan digunakan untuk segmentasi kata
    segmenter="twitter", 
    
    # korpus yang digunakan untuk statistik kata yang akan digunakan untuk koreksi ejaan
    corrector="twitter", 
    
    unpack_hashtags=True,  # melakukan segmentasi kata pada hashtag
    unpack_contractions=True,  # membuka kontraksi (can't -> can not)
    spell_correct_elong=False,  # koreksi ejaan untuk kata yang diperpanjang
    
    tokenizer=SocialTokenizer(lowercase=True).tokenize,
    
    # daftar kamus, untuk mengganti token yang diekstrak dari teks
    dicts=[emoticons]
)
```

```python
# panggil ekphrasis

def bersih_data(text):
    return " ".join(text_processor.pre_process_doc(text))
```

```python
# Fungsi untuk menghapus karakter non-ASCII
def non_ascii(text):
    return text.encode('ascii', 'replace').decode('ascii')

# Fungsi untuk menghapus spasi yang berlebihan
def remove_space_alzami(text):
    return " ".join(text.split())

# Fungsi untuk menghapus emoji
def remove_emoji_alzami(text):
    return ' '.join(re.sub("([x#][A-Za-z0-9]+)", " ", text).split())

# Fungsi untuk menghapus tab, newline, dan karakter Unicode lainnya
def remove_tab(text):
    return text.replace('\\t', " ").replace('\\n', " ").replace('\\u', " ").replace('\\', "")

# Fungsi untuk mengganti semua whitespace dengan satu spasi
def remove_tab2(text):
    return re.sub('\s+', ' ', text)

# Fungsi untuk menghapus 'RT' yang sering muncul di Twitter
def remove_rt(text):
    return text.replace('RT', " ")

# Fungsi untuk menghapus mention (@username) dan URL
def remove_mention(text):
    return ' '.join(re.sub("([@#][A-Za-z0-9]+)|(\w+:\/\/\S+)", " ", text).split())

# Fungsi untuk menghapus URL
def remove_incomplete_url(text):
    return text.replace("http://", " ").replace("https://", " ")

# Fungsi untuk menghapus karakter tunggal (huruf)
def remove_single_char(text):
    return re.sub(r"\b[a-zA-Z]\b", "", text)

# Fungsi untuk menghapus titik ganda (excessive dots)
def remove_excessive_dot(text):
    return text.replace('..', " ")

# Fungsi untuk mengganti strip (-) dengan spasi
def change_stripe(text):
    return text.replace('-', " ")

# Fungsi untuk mengubah teks menjadi huruf kecil
def lower(text):
    return text.lower()

# Fungsi untuk menghapus whitespace di awal dan akhir teks
def remove_whitespace_LT(text):
    return text.strip()

# Fungsi untuk mengubah whitespace berlebihan menjadi satu spasi
def remove_whitespace_multiple(text):
    return re.sub('\s+', ' ', text)

# Fungsi untuk menghapus tanda baca
def remove_punctuation(text):
    remove = string.punctuation
    remove = remove.replace("_", "")  # jangan hapus tanda hubung
    pattern = r"[{}]".format(remove)  # buat pola
    return re.sub(pattern, "", text)
```

```python
# hapus untuk <>
def remove_number_eks(text):
    return text.replace('<number>'," ")

def remove_angka(text):
    return re.sub(r"\d+", "", text) 

def remove_URL_eks(text):
    return text.replace('URL'," ").replace('url'," ")

def space_punctuation(text):
    return re.sub('(?<! )(?=[.,!?()])|(?<=[.,!?()])(?! )', r' ', text)
```

```python
i = 0
final_string = []
s = ""
for text in df['Tweet'].values:
    filteredSentence = []
    EachReviewText = ""
    proc = remove_rt(text)
    proc = lower(proc)
    proc = change_stripe(proc)
    proc = remove_emoji_alzami(proc)
    proc = remove_tab(proc)
    proc = remove_tab2(proc)
    proc = non_ascii(proc)
    proc = remove_incomplete_url(proc)
    proc = remove_excessive_dot(proc)
    proc = remove_whitespace_LT(proc)
    proc = remove_whitespace_multiple(proc)
    proc = remove_single_char(proc)
    proc = space_punctuation(proc)
    proc = remove_punctuation(proc)
    proc = remove_space_alzami(proc)
    proc = bersih_data(proc)
    proc = remove_number_eks(proc)
    proc = remove_angka(proc) 
    proc = remove_URL_eks(proc)
    EachReviewText = proc
    final_string.append(EachReviewText)
```

```python
df["step01"] = final_string
```

```python
normalized_word = pd.read_csv('kamus_clean.csv')
```

```python
normalized_word_dict = {}

for index, row in normalized_word.iterrows():
    if row[0] not in normalized_word_dict:
        normalized_word_dict[row[0]] = row[1] 

def normalized_term(document):
    return [normalized_word_dict[term] if term in normalized_word_dict else term for term in document]
```

```python
df_new['final_tokens'] = df_new['tokens'].apply(normalized_term)
```
